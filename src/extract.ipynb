```python
import pandas as pd
import csv
import numpy as np
```

```python
raw_m = pd.read_csv('appstore_games.csv')
raw_s = pd.read_csv('steam.csv')

# mobile: [name, age, developer, genre, price, date, rating(0-5), number of ratings, English,]
#         [in-app purchase]
# steam:  [name, age, developer, genre, price, date, rating(0,5), number of ratings, English,]
#         [publisher, median playtime, owners, platforms, categories, tags]
```

```python
print(raw_m)
mobile = raw_m[:]
```

```python
# delete not-game
lack_m = set()
for num, item in enumerate(mobile['Primary Genre']):
    if (item != 'Games'):
        lack_m.add(num)
mobile = mobile.drop(lack_m)

# modify age
for item in list(mobile['Age Rating']):
    item = int(item[0:-1])

# add English
english = list()
for item in list(mobile['Languages']):
    if type(item) == float or 'EN' in item:
        english.append(1)
    else: 
        english.append(0)
mobile.insert(2,'english', english)

# delete no-use attribute
mobile = mobile.drop(['URL','ID','Subtitle','Icon URL','Description','Size','Primary Genre','Current Version Release Date','Languages'], axis=1)

print(mobile)
```

```python
mobile.to_csv('mobile_modified.csv')
```

```python
print(raw_s)
steam = raw_s[:]
```

```python
# add average_ratings, rating_count
average_ratings = list()
rating_count = list()
for num in range(len(steam)):
    count = steam['positive_ratings'][num]+steam['negative_ratings'][num]
    rating_count.append(count)
    average_ratings.append(raw_s['positive_ratings'][num]*5/count)
steam.insert(11,'average_ratings', average_ratings)
steam.insert(12,'rating_count', rating_count)

# delete no-use attribute
steam = steam.drop(['appid','achievements','average_playtime','positive_ratings','negative_ratings'], axis=1)

print(steam)
```

```python
steam.to_csv('steam_modified.csv')
```
